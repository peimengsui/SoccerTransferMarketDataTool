```python

def get_players_data(players, rows,index,year):
    '''helper function to parse player data on who were transferred in the English premier league '''
    if year ==2014:
        club_list = ['Arsenal FC','Aston Villa','Burnley FC','Chelsea FC','Crystal Palace','Everton FC','Hull City',
                    'Leicester City','Liverpool FC','Manchester City','Manchester United','Newcastle United',
                     'Queens Park Rangers','Southampton FC','Stoke City','Sunderland AFC','Swansea City',
                     'Tottenham Hotspur','West Bromwich Albion','West Ham United']
    elif year == 2015:
        club_list = ['Chelsea FC', 'Manchester City','Arsenal FC','Manchester United','Tottenham Hotspur','Liverpool FC',
                 'Southampton FC','Swansea City','Stoke City','Crystal Palace','Everton FC','West Ham United',
                  'West Bromwich Albion','Leicester City','Newcastle United','Sunderland AFC','Aston Villa',
                  'AFC Bournemouth','Watford FC','Norwich City']
    elif year == 2016:
        club_list = ['Leicester City','Arsenal FC','Tottenham Hotspur','Manchester City',
             'Manchester United','Southampton FC','West Ham United','Liverpool FC',
            'Stoke City','Chelsea FC','Everton FC','Swansea City','Watford FC',
            'West Bromwich Albion','Crystal Palace','AFC Bournemouth','Sunderland AFC',
             'Burnley FC','Middlesbrough FC','Hull City']
    for row in rows[1:]:
        players_data = { }
        for (i,item) in  enumerate(row.find_all('td')):
                if i == 0:
                    parts = item.find('a')
                    players_data['p_name'] = parts.get('title')
                    #parts = item.find('a').get('href').split('/')
                    players_data['handle'] = parts.get('href').split('/')[1]
                    players_data['id'] = parts.get('id')
                if i == 1:
                    players_data['age'] = item.get_text()
                if i == 3:
                    players_data['position'] = item.get_text()
                if i == 5:
                    players_data['transfer_fee_pounds'] = item.get_text()
                if index%2 == 0:
                    players_data['transfer_status'] = 'in'
                else:
                    players_data['transfer_status'] = 'out'
                players_data['club'] = club_list[int(index/2)]
                
        players.append(players_data)
    
```

```python
def fetch_data(players, year):
    ''' given a particular year , fetch transferred player data from English premier league'''
    url = 'http://www.transfermarkt.co.uk/premier-league/sommertransfers/wettbewerb/GB1/saison_id/%s.html' % (year)
    r = requests.get(url,headers={"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36"})
    html_doc = r.text
    soup = BeautifulSoup(html_doc)
    table= soup.find_all('div',{'class':"responsive-table"})
    for i in range(0,39):
        rows = table[i].find_all('tr')
        get_players_data(players, rows,i,year)
    #rows = table.find_all('tr' , {'class':"dunkel"})
    #get_players_data(players, rows[1:])
```

```python
def extend_data(players, new_players, year):
    for player in players:

        url = 'http://www.transfermarkt.co.uk/%s/profil/spieler/%s' % (player['handle'],player['id'])
        print(player['handle'])
        r = requests.get(url,headers={"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36"})
        html_doc = r.text
        soup = BeautifulSoup(html_doc)
        try:
            table = soup.find('table' , {'class':"items"})
            row = table.find_all('tr')[3]
        except Exception:
            pass
        
        for (i,item) in  enumerate(row.find_all('td',{'class':"zentriert"})):
            if i == 0:
                if item.get_text() != '-':
                    player['matches'] = item.get_text()
                else:
                    player['matches'] = 0
            if i == 1:
                if item.get_text() != '-':
                    player['goals'] = item.get_text()
                else:
                    player['goals'] = 0
            if i == 2:
                if item.get_text() != '-':
                    player['assists'] = item.get_text()
                else:
                    player['assists'] = 0
            if i == 3:
                if item.get_text() != '-':
                    player['minutes_per_goal'] = item.get_text()
                else:
                    player['minutes_per_goal'] = 0
            if i == 4:
                if item.get_text() != '-':
                    player['minutes'] = item.get_text()
                else:
                    player['minutes'] = 0
        new_players.append(player)
```
